```python
import pandas as pd 
from matplotlib import pyplot as plt
import numpy as np
ToyotaCorolla=pd.read_csv("ToyotaCorolla.csv")
ToyotaCorolla.head()
```

```python
print(ToyotaCorolla.columns.values)
```

```python
plt.scatter(ToyotaCorolla.loc[:,'Age'], ToyotaCorolla.loc[:,'Price'])
plt.xlabel("Age")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(ToyotaCorolla.loc[:,'KM'], ToyotaCorolla.loc[:,'Price'])
plt.xlabel("KM")
plt.ylabel("Price")
plt.show()
```

```python
select_index= ['Age','KM']
ToyotaCorolla_x = ToyotaCorolla.loc[:,select_index] #2D PREDICTORS ARRAY
```

```python
ToyotaCorolla_y=ToyotaCorolla.loc[:,'Price'] 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_valid,y_train,y_valid =\
train_test_split(ToyotaCorolla_x,ToyotaCorolla_y, test_size=0.2,random_state=77)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
np.set_printoptions(precision=4, suppress=True) 
print('%.4f' %model.intercept_) 
```

```python
print(model.coef_) 
```

```python
#20440.3509 - 154.6958 * Age - 0.0154 * KM
```

```python
y_predict = model.predict(x_valid)
```

```python
y_new_predict = model.predict([[48,41636]])
```

```python
e = y_valid - y_predict
mae = np.mean(np.abs(e))
print(mae)
```
